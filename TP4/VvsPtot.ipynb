```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path and sheet names
file_path = "/Users/ludovica/Documents/Bachelor/BA6/TP GC/TP 4/Data.xlsx"
sheet_names = ["no.4", "no.5", "no.6", "no.7", "unknown"]
custom_labels = ["Particle 4", "Particle 5", "Particle 6", "Particle 7", "Particle Unknown"]  # Define custom labels

# Create a plot
plt.figure(figsize=(8, 6))

# Define colors for different sheets
colors = ['b', 'g', 'r', 'c', 'm']

# Loop through sheets and plot data
for sheet, color, label in zip(sheet_names, colors, custom_labels):
    df = pd.read_excel(file_path, sheet_name=sheet)
    flow_rate = df.iloc[:, 0]  # First column (assuming Flow Rate)
    pressure_loss = df.iloc[:, 1]  # Second column (assuming Pressure Loss)
    plt.plot(flow_rate, pressure_loss, marker='o', markersize=4, linestyle='-', linewidth=1, color=color, label=label)

# Labels and title
plt.xlabel("V' [m/s]")  
plt.ylabel("$\Delta p_f$  [Pa] ")   
plt.title("Pressure Drop vs. Flow Rate")
plt.legend()
plt.grid(True)
plt.show()
```
