```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Define the file path and sheet names
file_path = "/Users/ludovica/Documents/Bachelor/BA6/TP GC/TP 4/Data.xlsx"
save_folder = "/Users/ludovica/Documents/Bachelor/BA6/TP GC/TP 4/Graphs"
sheet_names = ["no.4", "no.5", "no.6", "no.7", "unknown"]
custom_labels = ["Particle 4", "Particle 5", "Particle 6", "Particle 7", "Particle Unknown"] 

# Ensure save folder exists
os.makedirs(save_folder, exist_ok=True)

# Define colors for different sheets
colors = ['b', 'g', 'r', 'c', 'm']

# Create a plot
plt.figure(figsize=(10, 6), dpi=600)  # High-quality image

# Loop through sheets and plot data
for sheet, color, label in zip(sheet_names, colors, custom_labels):
    df = pd.read_excel(file_path, sheet_name=sheet)
    flow_rate = df.iloc[:, 0]  # First column (Flow Rate)
    pressure_loss = df.iloc[:, 1]  # Second column (Pressure Loss)
    plt.plot(flow_rate, pressure_loss, marker='o', markersize=5, linestyle='-', linewidth=1.5, color=color, label=label)

# Labels and title
plt.xlabel("Flow Rate [m/s]", fontsize=12)
plt.ylabel("Pressure Drop [Pa]", fontsize=12)
plt.title("Pressure Drop vs. Flow Rate", fontsize=14, weight='bold')

# Customize grid and styling
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.gca().set_facecolor('#f9f9f9') 
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Add legend
plt.legend(fontsize=10)
plt.grid(False)

# Save the figure
save_path = os.path.join(save_folder, "VvsPtot.png")
plt.savefig(save_path, format='png', dpi=600, bbox_inches='tight')

# Show the plot
plt.show()

print(f"Plot saved successfully at: {save_path}")
```
