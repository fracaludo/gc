```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
from scipy.optimize import curve_fit

# File paths
file_path = "/Users/ludovica/Documents/Bachelor/BA6/TP GC/TP 4/Data.xlsx"
save_folder = "/Users/ludovica/Documents/Bachelor/BA6/TP GC/TP 4/Graphs"
sheet_name = "no.5"
label = "Particle 5"

# Ensure save folder exists
os.makedirs(save_folder, exist_ok=True)

# Constants
mu = 0.00089  # Fluid viscosity (Pas) 
rho = 997  # Fluid density (kg/m) 
L = 0.193  # Packed bed height (m)

# Two different porosity values
epsilon_1 = 0.395  
epsilon_2 = 0.464  
d_p = 0.002  # Particle diameter (m) 

# Define the Ergun fit equation
def ergun_eq(flow_rate, a, b):
    return a * flow_rate + b * flow_rate**2

# Read the data
df = pd.read_excel(file_path, sheet_name=sheet_name)
flow_rate = df.iloc[:, 0]  # First column (Flow Rate)
pressure_loss = df.iloc[:, 1]  # Second column (Pressure Loss)

# Fit Ergun's equation to get coefficients a and b
popt, _ = curve_fit(ergun_eq, flow_rate, pressure_loss)
a_fit, b_fit = popt

# Theoretical Ergun equation
def ergun_theoretical(flow_rate, epsilon, d_p):
    term1 = (150 * (1 - epsilon) ** 2 * mu * flow_rate) / (d_p**2 * epsilon ** 3)
    term2 = (1.75 * (1 - epsilon) * rho * flow_rate**2) / (d_p * epsilon ** 3)
    return (term1 + term2) * L  # Multiply by bed height

# Generate theoretical Ergun values
ergun_values_eps1 = ergun_theoretical(flow_rate, epsilon_1, d_p)
ergun_values_eps2 = ergun_theoretical(flow_rate, epsilon_2, d_p)

# Calculate R-squared value
fitted_pressure_loss = ergun_eq(flow_rate, *popt)
residuals = pressure_loss - fitted_pressure_loss
ss_res = np.sum(residuals**2)
ss_tot = np.sum((pressure_loss - np.mean(pressure_loss))**2)
r_squared = 1 - (ss_res / ss_tot)

# Create a plot with both experimental data, Ergun fit, and theoretical Ergun equation
plt.figure(figsize=(10, 6), dpi=600)  

# Plot experimental data
plt.plot(flow_rate, pressure_loss, 'o', markersize=5, color='g', label="Experimental Data")

# Plot Ergun fit
plt.plot(flow_rate, fitted_pressure_loss, '-', color='darkorange', linewidth=1.5, 
         label=f"Ergun Fit: a={a_fit:.1f}, b={b_fit:.1f}, R={r_squared:.4f}")

# Plot theoretical Ergun equation
plt.plot(flow_rate, ergun_values_eps1, '--', color='red', linewidth=1.5, 
         label=f"Theoretical Ergun 1 (={epsilon_1}, d_p={d_p:.4f} m)")
plt.plot(flow_rate, ergun_values_eps2, '--', color='blue', linewidth=1.5, 
         label=f"Theoretical Ergun 2 (={epsilon_2}, d_p={d_p:.4f} m)")

# Labels and title
plt.xlabel("V' [m/s]", fontsize=12)
plt.ylabel("$\Delta p_f$  [Pa] ", fontsize=12)
plt.title(f"Pressure Drop vs. Flow Rate for {label}", fontsize=14, weight='bold')

# Customize grid and styling
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.gca().set_facecolor('#f9f9f9')  # Subtle background color
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Add legend
plt.legend(loc='upper left', fontsize=10)
plt.grid(False)

# Save the figure
save_path = os.path.join(save_folder, f"Comparison_{label.replace(' ', '_')}.png")
plt.savefig(save_path, format='png', dpi=600, bbox_inches='tight')

# Show the plot
plt.show()

print(f"Plot saved successfully at: {save_path}")
```
